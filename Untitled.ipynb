```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L, N = 500, 500

# values of lines
a = np.arange(L)
b = a*3.1
```

```python
def pos(N,array):
    return np.array([np.cos(array*(2*np.pi)/N), np.sin(array*(2*np.pi)/N)]).T
```

```python
a_pos, b_pos = pos(N,a), pos(N,b)
```

```python
a_pos.shape
```

```python
for i in range(L):
    plt.plot(a_pos[i], b_pos[i], 'ro-', linewidth=0.1, markersize=0.1)
plt.show()
```

```python
L, N = 500, 500
a = np.arange(L)
b = a*103.1

a_pos, b_pos = pos(N,a), pos(N,b)
for i in range(L):
    x1, y1 = a_pos[i]
    x2, y2 = b_pos[i]
    plt.plot((x1,x2),(y1,y2),'ro-',linewidth=0.1, markersize=0.1)
plt.show()
```
